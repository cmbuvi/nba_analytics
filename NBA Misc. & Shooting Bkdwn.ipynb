```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_16 = pd.read_excel("misc_reg_agg.xlsx")
fltr16 = df_16[df_16['Year'] != 2016].index
df_16.drop(fltr16, inplace=True)
# df_16.hist('W/L')

df_17 = pd.read_excel("misc_reg_agg.xlsx")
fltr17 = df_17[df_17['Year'] != 2017].index
df_17.drop(fltr17, inplace=True)
# df_17.hist('W/L')

df_18 = pd.read_excel("misc_reg_agg.xlsx")
fltr18 = df_18[df_18['Year'] != 2018].index
df_18.drop(fltr18, inplace=True)
# df_18.hist('W/L')

df_19 = pd.read_excel("misc_reg_agg.xlsx")
fltr19 = df_19[df_19['Year'] != 2019].index
df_19.drop(fltr19, inplace=True)
# df_19.hist('W/L')

df_20 = pd.read_excel("misc_reg_agg.xlsx")
fltr20 = df_20[df_20['Year'] != 2020].index
df_20.drop(fltr20, inplace=True)
# df_20.hist('W/L')

df = pd.read_excel("misc_reg_agg.xlsx")
# df.hist('W')
df.hist('W/L')
df.head()
```

```python
corrMat = df.corr(method='pearson')
fig, ax = plt.subplots(figsize = (20,20))
ax = sns.heatmap(corrMat, cbar=0, linewidths=0.5,vmax=1, vmin=0, square=True, cmap="YlGn", annot=True)
ax.figure.savefig("reg_heatmap.png")
plt.show()
```

```python
win_corr = corrMat['W/L']
win_corr.sort_values(ascending = False)
```

```python
misc_df = pd.read_excel("misc_reg_agg.xlsx")
perc_df = pd.read_excel("team_per_100_agg.xlsx")
shot_df = pd.read_excel("shooting_reg_agg.xlsx")

agg_df = shot_df.merge(perc_df)
agg_df = agg_df.merge(misc_df)
# agg_df.hist('W/L')
# agg_df.head()

agg_corrMat = agg_df.corr(method='pearson')
fig1, ax1 = plt.subplots(figsize = (30,30))
ax1 = sns.heatmap(agg_corrMat, cbar=0, linewidths=0.5,vmax=1, vmin=0, square=True, cmap="BuGn", annot=True)
ax1.figure.savefig("agg_heatmap.png")
plt.show()
```

```python
agg_corr = agg_corrMat['MOV']

agg_corr.sort_values(ascending = False)
```
